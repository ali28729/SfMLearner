```python
import numpy as np
from matplotlib import pyplot as plt
import cv2

img = cv2.imread('misc/tests/poop3DepthLatest.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img)
print(img.shape)
h, w = img.shape[:2]
```

```python
fig = plt.figure(figsize=(42, 5))
columns = 10
rows = 1
upper = .95
lower = .80

for x in range(0, 10):
    bins = img[50:-120,x*(w//10):(x+1)*(w//10)]
    print(bins.shape)
    hist = np.bincount(bins.ravel(),minlength=256)
    c_hist = np.cumsum(hist)
    u = -1
    l = -1
    total = 0
    count = 0
    for i in range(255,-1,-1):
        if c_hist[i] < c_hist[-1] * upper:
            total += hist[i] * i
            count += hist[i]
        if c_hist[i] < c_hist[-1] * lower:
            break
    avg = total/count
    print(avg)        
    
    fig.add_subplot(rows, columns, x+1)
    plt.imshow(bins)
plt.show()
```
